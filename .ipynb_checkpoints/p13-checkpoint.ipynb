Algo simple que se ha pensado para identificar la sonrisa es determinar una zona en la imagen que tenga varios pixeles que correspondan al color blanco. (La imagen ha sido previamente seleccionada para los propósitos)

Trabajar en blanco y negro no es la mejor opción ya que la cara quedará marcada como color blanco y esto no ayuda a la idea que se tiene para identificar la sonrisa, por lo que se procede a usar escala de grises.

```python
from PIL import ImageOps
gray = ImageOps.grayscale(imagen)
gray
```

El siguiente paso es marcar los pixeles que quedan más cerca del blanco, es decir cuyo valor es cercano a 255.

```python
from PIL import ImageOps
gray = ImageOps.grayscale(imagen)
P = gray.load()
ancho, altura = gray.size
for fila in range(altura):
    for columna in range(ancho):
        if P[columna, fila] < 200:
            P[columna, fila] = 0 # será negro
gray
```

```python
from PIL import ImageDraw
import numpy as np

draw = ImageDraw.Draw(gray)
m=20
n=40
for i in range(0,altura-n,n):
    for j in range(0,ancho-m,m):
        crop_rectangle = (j, i, j+m, i+n)
        rectangle = gray.crop(crop_rectangle)
        rectangle_rgb = np.array(rectangle)
        negros = 0
        for k in range(m):
            for l in range(n):
                if rectangle_rgb[l,k,0]<200 and rectangle_rgb[l,k,1]<200 and rectangle_rgb[l,k,2]<200:
                    negros = negros + 1
        #print(blancos)
        if negros < 100:
            draw.rectangle([j,i,j+m,i+n], outline="red")
            c=c+1
print(c)
gray
```
